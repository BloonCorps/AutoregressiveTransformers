```python
import numpy as np
import pickle
import math
from sys import exit
import matplotlib.pyplot as plt
from sklearn import datasets
import argparse
```

$\Delta G \degree \prime$

```python
X, label = datasets.make_moons(n_samples = 10000, noise = 0.05)
X = X[X[:, 0].argsort()]
```

```python
plt.scatter(X[:, 0], X[:, 1], s=0.4)
plt.title("Two Moons Dataset 2D Distribution")
plt.xlabel("$X$")
plt.ylabel("$Y$")
```

```python
plt.hist(X[:, 0], bins = 100)
plt.title("Marginal X Distribution")
plt.show()
```

```python
plt.hist(X[:, 1], bins = 100)
plt.title("Marginal Y Distribution")
plt.show()
```

```python
markers = np.linspace(-1.2, 2.3, num = 30)
```

```python
index_list = []
mark_indx = 0
current_marker = markers[mark_indx]
for indx in range(0, 10000, 1):
    if X[indx, 0] > current_marker:
        index_list.append(indx)
        mark_indx += 1
        current_marker = markers[mark_indx]
```

```python
print(index_list)
```

```python
pairs = []
```

```python
for index in range(0, len(index_list)-1, 1):
    pairs.append((index_list[index], index_list[index+1]))
```

```python
print(pairs)
```

```python
 str(round(markers[0],2))
```

```python
from matplotlib.pyplot import figure

index = 0
for pair in pairs:
    start = pair[0]
    stop = pair[1]

    fig, axs = plt.subplots(2, figsize=(6, 9), dpi=100)

    axs[0].set_title("Two Moons Dataset")
    axs[0].scatter(X[:, 0], X[:, 1], s=0.4)
    axs[0].axvline(x=markers[index], color='r')
    axs[0].axvline(x=markers[index+1], color='r')

    axs[1].hist(X[start:stop, 1], range=[-0.60, 1.1], bins=150)
    axs[1].set_title("$p(y \ |  \ $" + str(round(markers[index], 2)) + "$ < x < $" + str(round(markers[index+1], 2)) + ")")
    
    #plt.subplots_adjust(top = 2, bottom = 0)
    index = index + 1
    plt.show()
    fig.savefig("index" + str(index) + ".png")
```
